# Apply ROIs, PCA

```python
import numpy as np
from nilearn import image
from nilearn.image import load_img
from nilearn.masking import apply_mask
from nilearn.plotting import plot_img, plot_roi, plot_anat
import os
import glob
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Get and list all of the ROIs
homeDir = '/home/micah/Documents/act-r-verification-master/'
os.chdir(homeDir)
ROIs = glob.glob("ROIs/*.nii.gz")
#print(ROIs)
header = ['left_imaginal', 'right_imaginal', 'left_retrieval', 'left_manual', 'right_visual', 'left_visual', 'right_retrieval', 'right_manual']
```

```python
# Get and list all of the subjects
subjDir = '/projects/BGA/Stroop2/ds000164_R1.0.1'
os.chdir(subjDir)
SUBJs = glob.glob("sub-*")
#print(SUBJs)
```

```python
# for every subject
for x in range(0,len(SUBJs)):
    img = image.load_img(SUBJs[x] + "/func/swa" + SUBJs[x] + "_task-stroop_bold.nii")
    #print(SUBJs[x] + "/func/swa" + SUBJs[x] + "_task-stroop_bold.nii")
    
    # open subject file and write headers
    with open(homeDir + SUBJs[x] + ".csv", "w") as f:
        f.write(",".join(header) + '\n')
    
        # for every ROI
        pcaList = [[] for x in range(0,len(ROIs))]
        for y in range(0,len(ROIs)):
            mask = image.load_img(homeDir + "/" + ROIs[y])
            #print(homeDir + ROIs[y])
            curROI = apply_mask(img, image.index_img(mask, 0))

            # show figure of each
            #fig = plt.figure()
            #plt.plot(curROI)
            #fig.set_size_inches(20, 5)
            #plt.show()

            ## PCA
            # normalize data
            scaler = StandardScaler()
            scaler.fit(curROI)
            normalized = scaler.transform(curROI)

            # print normalized data
            #fig = plt.figure()
            #plt.plot(normalized)
            #fig.set_size_inches(20, 5)

            pca = PCA(n_components=1)
            pca.fit(normalized)

            # project data
            # OUTPUT THIS TO THE FILE
            projected = pca.transform(curROI)
            pcaList[y] = projected;

            var_projected = np.sum(np.var(projected, axis=0))
            var_original = np.sum(np.var(curROI, axis=0))

            # percentage of variance explained
            varExplained = var_projected / var_original

            if varExplained > .30:
                print(SUBJs[x] + "_" + ROIs[y]+ ": fuck yeah")
            else:
                #write in file if so
                print(SUBJs[x] + "_" + ROIs[y]+ ": fuck NOOOOOOOOOOOOOOOOOOOOOOOOOOOOOOO")
         
        # for each timepoint
        for x in range(0,len(pcaList[0][:])):
            row = []
            for i in range(0,len(ROIs)):
                row.append(pcaList[i][x][0])
            f.write(",".join([str(i) for i in row]) + '\n')
```
